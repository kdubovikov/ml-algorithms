```python
from sklearn.datasets import make_classification
import numpy as np

x, y = make_classification(n_features=3, n_informative=2, n_redundant=1)
y = np.asarray(y, dtype='bool')
```

Lets create a simple dataset.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x1 = x[y]
x2 = x[~y]
ax.plot(x1[:, 0], x1[:, 1], x1[:, 2], 'o', alpha=0.5)
ax.plot(x2[:, 0], x2[:, 1], x2[:, 2], 'o', alpha=0.5);
```

We can use eigendecomposition to compute rank $n$ PCA.

$$\mathbf{X} = \underset{i, j \le n}{\mathbf{V \Lambda} \mathbf{V}^T}$$

```python
cov = np.cov(x, rowvar=False)
e_val, e_vec = np.linalg.eig(cov) 

order = 2
sorted_idx = e_val[::-1].argsort()
e_val = e_val[sorted_idx]
e_vec = e_vec[:, sorted_idx]

e_val = e_val[:order]
e_vec = e_vec[:order]
```

```python
x_t = e_vec.dot(x.T).T

fig = plt.figure()
ax = fig.add_subplot(111)

x1 = x_t[y]
x2 = x_t[~y]
ax.plot(x1[:, 0], x1[:, 1], 'o', alpha=0.5)
ax.plot(x2[:, 0], x2[:, 1], 'o', alpha=0.5);
```
